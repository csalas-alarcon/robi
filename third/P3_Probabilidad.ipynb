```python
# Ejercicio 3
import random
import pandas as pd
import matplotlib.pyplot as plt

# Simular 1000 lanzamientos
n_lanzamientos = 1000
resultados = [random.randint(1, 6) for _ in range(n_lanzamientos)]

# Tabla de frecuencias
df = pd.Series(resultados).value_counts().sort_index().reset_index()
df.columns = ['Cara', 'Frecuencia']
df['Porcentaje (%)'] = (df['Frecuencia'] / n_lanzamientos) * 100
print(df)

# Visualización
plt.bar(df['Cara'], df['Frecuencia'], color='lightgreen', edgecolor='black')
plt.title('Simulación de Lanzamiento de un Dado')
plt.xlabel('Cara del Dado')
plt.ylabel('Frecuencia')
plt.xticks(range(1, 7))
plt.show()
```

```python
# Ejercicio 9
# La primera carta puede ser cualquiera (probabilidad 1)
# La segunda extracción debe coincidir con la primera (1 carta entre 52)
prob_repetir = 1 * (1 / 52)

print(f"Probabilidad de repetir la carta: {prob_repetir:.6f}")
```

```python
# Ejercicio 10
bolas = [1, 2, 3, 4, 5]
casos_posibles = len(bolas) ** 2
casos_favorables = 0

for bola1 in bolas:
    for bola2 in bolas:
        if (bola1 + bola2) % 2 != 0:
            casos_favorables += 1

prob = casos_favorables / casos_posibles
print(f"Probabilidad de que la suma sea impar: {prob}")
```

```python
# Ejercicio 11
prob_x = 0.4
prob_union = 0.6

# Ecuación: 0.6 = 0.4 + P(Y) - 0.4*P(Y)
# 0.2 = 0.6 * P(Y)
prob_y = (prob_union - prob_x) / (1 - prob_x)

print(f"Probabilidad del suceso Y: {prob_y:.4f}")
```

```python
# Ejercicio 12
# Probabilidad de elegir cualquiera de las 3 urnas
p_urna = 1 / 3

# Probabilidades de sacar bola roja en cada urna
p_r_dado_a = 3 / 8
p_r_dado_b = 2 / 3  # 2 rojas y 1 negra
p_r_dado_c = 2 / 5

# Probabilidad total de sacar bola roja
p_r = (p_r_dado_a * p_urna) + (p_r_dado_b * p_urna) + (p_r_dado_c * p_urna)

# Teorema de Bayes
p_a_dado_r = (p_r_dado_a * p_urna) / p_r

print(f"Probabilidad de que la bola sea de la urna A: {p_a_dado_r:.4f}")
```

```python
# Ejercicio 13
# Datos de puertas por grupo
puertas_a = 380
puertas_b = 270
puertas_c = 350
total_puertas = puertas_a + puertas_b + puertas_c

# Cálculo de puertas defectuosas
defectuosas_a = puertas_a * 0.04
defectuosas_b = puertas_b * 0.03
defectuosas_c = puertas_c * 0.06
total_defectuosas = defectuosas_a + defectuosas_b + defectuosas_c

# Probabilidad total
prob_defectuosa = total_defectuosas / total_puertas

print(f"Probabilidad de puerta defectuosa: {prob_defectuosa:.4f}")
```

```python
# Ejercicio 14
# Probabilidades de compra (al primero se le compra 5 veces más)
p_prov2 = 1 / 6
p_prov1 = 5 / 6

# Probabilidades de rechazo por proveedor
p_rechazo_prov1 = 0.20
p_rechazo_prov2 = 0.10

# Probabilidad total de encontrar una papa de rechazo
p_rechazo = (p_prov1 * p_rechazo_prov1) + (p_prov2 * p_rechazo_prov2)

# Teorema de Bayes: P(Proveedor 2 | Rechazo)
p_prov2_dado_rechazo = (p_prov2 * p_rechazo_prov2) / p_rechazo

print(f"Probabilidad de que la papa sea del proveedor 2: {p_prov2_dado_rechazo:.4f}")
```
