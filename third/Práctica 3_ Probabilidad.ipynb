```python
try:
    import texttable # librería diseñada para generar tablas ASCII (tablas hechas con texto)
except ImportError:
    !pip install texttable
    import texttable
from texttable import Texttable # Textable (T en mayúscula) es una clase dentro del paquete textable.
# Función para simular lanzamientos de moneda y obtener resultados
def lanzamiento_moneda(n):
    resultados = [random.choice(['cara', 'cruz']) for _ in range(n)] # _: necesito un contador para el bucle, pero no voy a usar su valor para nada.
    return resultados

# Simulación de lanzar la moneda 100 veces
resultados = lanzamiento_moneda(100)

# Contar la frecuencia de cada resultado
cara_count = resultados.count('cara')
cruz_count = resultados.count('cruz')



# Tabla Resumen
t1 = Texttable()
t1.add_rows([['Resultado','frecuencia', 'Porcentaje' ],
           ['cara',	cara_count, cara_count/(cara_count+cruz_count)*100],
           ['cruz',	cruz_count, cruz_count/(cara_count+cruz_count)*100]])
print(t1.draw())

# Generar histograma
plt.bar(['Cara', 'Cruz'], [cara_count, cruz_count], color = ['blue', 'salmon'])
plt.title('Diagrama de barras de Lanzamiento de Moneda')
plt.xlabel('Resultado')
plt.ylabel('Frecuencia')
plt.show()
```

## ESPACIO MUESTRAL

Vamos a ver un código python para determinar, por enumeración, el espacio muestral del experimento consistente en lanzar una moneda dos veces.

```python
# Producto cartesiano: Conjunto A con "n" elementos y realizas un experimento "k" veces el número total de resultados es: n^k. (Para casos independientes)
posibles = {"cara","cruz"}
espacio_muestral = set()
for tiro1 in posibles:
    for tiro2 in posibles :
        salida = (tiro1 , tiro2 )
        espacio_muestral.add( salida )
print(espacio_muestral)
```

# UNIÓN / INTERSECCIÓN

Vamos a ver un código python para determinar, por enumeración, la unión y la intersección de dos conjuntos

UNIÓN

```python
# Creación de los conjuntos (los duplicados se eliminan solos)
set1 = set([1, 1, 2, 3, 4, 5])
set2 = set([1, 5, 6, 7, 8, 9])
set3 = set([3, 4, 5, 3, 9, 10])
set4 = set([5, 7, 9, 10, 12, 14])
print("Elementos de los conjuntos originales:")
print(set1)
print(set2)
print(set3)
print(set4)

# Unión de todos los conjuntos en uno solo
# Puedes pasar múltiples conjuntos al método .union()
union_total = set1.union(set2, set3, set4)

print("\nUnión de todos los conjuntos:")
print(union_total)
```

INTERSECCIÓN

```python
# Creación de los conjuntos
setx = {"verde", "azul"}
sety = {"azul", "amarillo"}

print("Conjunto X:", setx)
print("Conjunto Y:", sety)

# Intersección: Solo los elementos que están en AMBOS conjuntos
resultado = setx.intersection(sety)

print("\nIntersección de los dos conjuntos:")
print(resultado)
```

## Definición de probabilídad.

Definición clásica: la probabilidad de un suceso $A$ es el cociente entre el número de casos favorables y el número de casos posibles.

Definición frecuentista (ley de grandes números): la probabilidad de un suceso es el valor límite de su frecuencia relativa al repetir indefinidamente la experimentación.

En teoría de probabilidad, podemos decir que dos eventos son independientes cuando la probabilidad de cada uno de ellos no se ve afecta porque el otro evento ocurra, es decir que no existe ninguna relación entre los eventos.

Se puede formular una "ley de grandes números" de la siguiente forma: si se repite un experimento aleatorio, bajo las mismas condiciones, un número ilimitado de veces; y si estas repeticiones son independientes la una de la otra, entonces la frecuencia de veces que un evento ocurra, convergerá a un número que es igual a la probabilidad de que ocurra en una sola repetición del experimento.

Para entender mejor la ley de grandes números, vamos a utilizar el ejemplo del lanzamiento de una moneda. En un primer ejemplo, la moneda va a tener la misma posibilidad de caer en cara o cruz. Mientras que en el segundo ejemplo, vamos a modificar la probabilidad de la moneda para que caiga cara solo en 1 de cada 6 veces.

MONEDA

```python
# Ejemplo ley de grandes números
# moneda p=1/2 cara=1 cruz=0
resultados = []
for lanzamientos in range(1,10000):
    tiradas = np.random.choice([0,1], lanzamientos) #Como se hace el promedio con #Mean, sale directamente el número de caras.
    caras = tiradas.mean()
    resultados.append(caras)

# graficamente
fig, ax = plt.subplots()
width = 8
height = 6
fig.set_size_inches(width, height)
df = pd.DataFrame({ 'tiradas' : resultados})
ax.plot(df)
xlabel = ax.set_xlabel("Número de lanzamientos")
ylabel = ax.set_ylabel ("frecuencia de caras")
plt.show()
```

DADO DE 6 CARAS

```python
# moneda p=1/6 sale=1 no sale=0
resultados = []
for lanzamientos in range(1,10000):
    lanzamientos = np.random.choice([0,1], lanzamientos, p=[5/6, 1/6])
    caras = lanzamientos.mean()
    resultados.append(caras)

# graficamente
fig, ax = plt.subplots()
width = 8
height = 6
fig.set_size_inches(width, height)
df = pd.DataFrame({ 'lanzamientos' : resultados})
ax.plot(df)
xlabel = ax. set_xlabel("Número de lanzamientos")
ylabel = ax.set_ylabel ("frecuencia de caras")
plt.show()
```

## EJEMPLO 1

Una empresa ofrece a sus clientes la posibilidad de solicitar un préstamo para un vehículo y un préstamo educativo. Un estudio muestra que la probabilidad de que un cliente solicite un préstamo para un vehículo es de 0,35, la probabilidad de que un cliente solicite un préstamo educativo es de 0,46 y la probabilidad de que un cliente solicite al menos uno de los préstamos es de 0,59. Halle la probabilidad de que un cliente solicite tanto un préstamo para un vehículo como un préstamo para estudios.

```python
# suceso A: cliente toma un prestamo para vehiculo
prob_vehicle = 0.35

# # suceso B: cliente toma un prestamo educaciónal 
prob_edu = 0.46

# probabilidad de que un cliente tomo uno de los dos (P(A ⋃ B))
prob_vehicle_or_edu = 0.59

# tLa probabilidad de que un cliente tome los dos (P(A ⋂ B))
# para eventos NO mutuamente exclusivos, P(A ⋃ B) = P(A) + P(B) - P(A ⋂ B) luego 
# P(A ⋂ B) = P(A) + P(B) - P(A ⋃ B) 
prob_vehicle_and_edu = prob_vehicle + prob_edu - prob_vehicle_or_edu

print(f'La probabilidad de que un cliente tome ambos prestamos es: {prob_vehicle_and_edu:.2f}')
```

## EJEMPLO 2

La paradoja del cumpleaños es un problema muy conocido en el campo de la teoría de probabilidad. Plantea las siguientes cuestión: ¿Cuál es la probabilidad de que, en un grupo de 23 personas elegidas al azar, al menos dos de ellas habrán nacido el mismo día del año? 
Nota: Excluyendo el 29 de febrero, se asume que los restantes 365 días de posibles cumpleaños son igualmente probables.

Se va a obtener la probabilidad de que en un grupo de 23 personas exista un cumpleaños duplicado. 

```python
# Ejemplo situación 2 La coincidencia de cumpleaños
prob = 1.0
asistentes = 23
for i in range(asistentes): #range 23 va de 0 a 22.
    prob = prob * (365-i)/365

print(f"Probabilidad de que compartan una misma fecha de cumpleaños es {1 - prob:.8f}")
```

## EJEMPLO 3

Se tienen dos urnas, y cada una de ellas contiene un número diferente de bolas blancas y rojas: Primera urna, U1: 3 bolas blancas y 2 rojas; Segunda urna, U2: 4 bolas blancas y 2 rojas.
Se realiza el siguiente experimento aleatorio: Se tira una moneda al aire y si sale cara se elige una bola de la primera urna, y si sale cruz de la segunda. ¿Cuál es la probabilidad de que salga una bola blanca?.

```python
def experimento_urnas():
    '''
    Función aleatoria que devuelve el resultado de simular el experimento *una vez*.
    El resultado es distinto cada vez que llamamos a la función.
    El resultado puede ser 'B' (bola blanca) o 'R' (bola roja)
    '''
    urna = random.randint(1,2) #se utiliza para generar un número entero aleatorio dentro de un rango específico que tú definas
    if urna == 1:
        #Escogemos una bola.
        return random.choice('BBBRR')
    else:
        #Escogemos una bola. Las bolas son [B,B,B,B,R,R]
        return random.choice('BBBBRR')

def es_bola_blanca(bola):
    '''
    Función que recibe como argumento un string, que puede ser
        "B" (representa una bola blanca)
        "R" (representa una bola roja)
    Devuelve True <=> el argumento es "B" (una bola blanca)
    '''
    if bola =='B':
        return True
    else:
        return False

random.seed(1) #Prueba a cambiar la semilla: la aproximación cambia 
N = 10000000 # prueba a cambiar el valor
muestra = [experimento_urnas() for _ in range(N)]
prob = sum(es_bola_blanca(e) for e in muestra)/N
print('Probabilidad aproximada de bola blanca:', prob)
```
